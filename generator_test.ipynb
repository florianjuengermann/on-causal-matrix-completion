```python
from src import generator as gen
import matplotlib.pyplot as plt
import numpy as np
```

## Limited MNAR

```python
rating_matrix, P = gen.getRatingAndPropensityMatrix(inv_scale=1)

plt.figure(figsize=(10, 10))
plt.imshow(P, cmap="binary"); plt.show()

# sample D:
D = np.random.binomial(1, P)
plt.figure(figsize=(6, 6))
plt.spy(D); plt.show()

# distribution of true vs revealed samples:
plt.hist(rating_matrix.flatten(), bins=100, alpha = 0.3, label="true samples")
plt.hist(rating_matrix[D==1].flatten(), bins=100, color = 'C0', label="revealed samples")
plt.grid(alpha=0.5)
plt.legend(loc="upper right");
plt.xlabel('ratings')
plt.ylabel('frequency');
```

## General MNAR

```python
rating_matrix, P, latent_movie_matrix = gen.getRatingAndPropensityMatrix_general(inv_scale=1)

plt.figure(figsize=(10, 10))
plt.imshow(P, cmap="binary")
plt.show();
# distribution of true vs revealed samples:
plt.hist(rating_matrix.flatten(), bins=100, alpha = 0.3)
plt.hist(rating_matrix[P==1].flatten(), bins=100, color = 'C0')
plt.xlabel('ratings')
plt.ylabel('frequency');
print(f"total: μ={np.mean(rating_matrix):.3f}")
print(f"revealed: μ={np.mean(rating_matrix[P==1]):.3f}")
```
